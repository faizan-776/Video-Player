### Step 2: Set up Tunnel (Ngrok)
Get your token from [https://dashboard.ngrok.com/get-started/your-authtoken](https://dashboard.ngrok.com/get-started/your-authtoken)

```python
from pyngrok import ngrok
import getpass

authtoken = getpass.getpass("Enter ngrok authtoken: ")
ngrok.set_auth_token(authtoken)
public_url = ngrok.connect(8000).public_url
print(f"\nBACKEND URL: {public_url}")
print("\nPaste this into your browser console:")
print(f"localStorage.setItem('BACKEND_URL', '{public_url}')")
```

### Step 3: Run the Backend
Note: Ensure you have uploaded the `backend/` folder to Colab or you are running this from VS Code's Colab extension with access to the files.

```python
import sys
import os
import uvicorn
import nest_asyncio

# 1. Allow nested event loops (required for Jupyter/Colab)
nest_asyncio.apply()

# 2. Correct paths for imports
root_dir = os.getcwd()
backend_dir = os.path.join(root_dir, 'backend')

if root_dir not in sys.path: sys.path.append(root_dir)
if backend_dir not in sys.path: sys.path.append(backend_dir)

try:
    from backend.main import app
    print("Imported app from backend.main")
except ImportError:
    from main import app
    print("Imported app from main")

# 3. Run uvicorn
config = uvicorn.Config(app, host="0.0.0.0", port=8000, loop="asyncio")
server = uvicorn.Server(config)
server.run()
```
